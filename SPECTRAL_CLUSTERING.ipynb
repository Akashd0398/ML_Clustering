```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset= pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:,[3,4]].values
#above step used to pick selevted rows or columns, values - shows the values 
```

```python
X
```

```python
from sklearn.cluster import SpectralClustering
SC = SpectralClustering(n_clusters = 2, assign_labels = 'discretize',random_state = 5 ).fit(X)
y_SC = SC.fit_predict(X)
```

```python
y_SC
```

```python
supervised = dataset
```

```python
supervised
```

```python
supervised['Cluster_group']=y_SC
#adding cluster group to the exeisting csv file
```

```python
supervised
```

```python
supervised.to_csv("cluster_SC.csv",index = False)
#saving new csv file
```

```python
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], 
                   fit_reg=False, legend=True, legend_out=True)
```
